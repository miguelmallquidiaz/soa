## Principios de SOA

### Estandarizacin de contratos de servicios

El contrato es un conjunto de documentos que va a hacer que otros principios se apliquen de la forma esperada.
Esta compuesto el contrato de servicio web mendiante el WSDL que es una estandarizacin de como el documento se van a ingresar los parametros de entrada y de igual forma que espera de salida, la nomenclatura se pone el nombre de la entidad verbo y sustantivo. Se comunica mediante el esquema XML y se aplica WS-Policy para temas de seguridad de como se va a acceder al servicio.

### Abstracin de servicios

El contrato de servicios contiene solamente la informacin esencial y adems no se va a esponer lgica de negocio.

### Acoplamiento dbil de servicios

Se enfoca en lograr un acoplamiento bajo para promover la independencia de servicios haciendo que sea fcil de mantener y reutilizar.

### Capacidad de descubrir el documento

Los servicios deben ser facilmente localizables y entendibles. Deben brindar la informacin suficiente del servicio para no tener servicios duplicados.

### Autonomia del servicio

Un servicio tiene su propio control de ejecucin para operar de manera independiente. Si el servicio A esta ejecutando en su entorno de ejecucin el servicio B cae, el servicio A no debe interumpirse.

### Servicio sin estado

Un servicio no debe guardar ningn tipo de informacin, todos los parametros de entrada y lo que se espera de salida deben proveenir directamente de la documentacin. 

### Componibilidad del servicio

Juntar servicios de bajo nivel para que sea un servicio de mayor nivel.

### Reutilizacin de servicios

El principal objetivo de SOA es construir servicios altamente reutilizables enfocado que se va a utilizar en varios aplicaciones y esto generar un valor ya que el mantenimiento sera ms fcil de implementar.

## SOAP VS REST VS grcp

- SOAP: Es un estndar para el intercambio de informacin utilizando XML
- REST: En cambio rest es una propuesta para crear servicios web, aunque no tiene estndares de seguridad tan avanzados como SOAP.
- GRCP: Es un marco de comunicacin diseada por Google, los datos se serealizan en formatos binarios usando Protocol Buffer, convierte la infomacin en formato binario lo que hace que trasmita ms rpido y consuma menos ancho de banda que otros formatos como JSON o XML.
