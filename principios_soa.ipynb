## Principios de SOA

### Estandarización de contratos de servicios

El contrato es un conjunto de documentos que va a hacer que otros principios se apliquen de la forma esperada.
Esta compuesto el contrato de servicio web mendiante el WSDL que es una estandarización de como el documento se van a ingresar los parametros de entrada y de igual forma que espera de salida, la nomenclatura se pone el nombre de la entidad verbo y sustantivo. Se comunica mediante el esquema XML y se aplica WS-Policy para temas de seguridad de como se va a acceder al servicio.

### Abstración de servicios

El contrato de servicios contiene solamente la información esencial y además no se va a esponer lógica de negocio.

### Acoplamiento débil de servicios

Se enfoca en lograr un acoplamiento bajo para promover la independencia de servicios haciendo que sea fácil de mantener y reutilizar.

### Capacidad de descubrir el documento

Los servicios deben ser facilmente localizables y entendibles. Deben brindar la información suficiente del servicio para no tener servicios duplicados.

### Autonomia del servicio

Un servicio tiene su propio control de ejecución para operar de manera independiente. Si el servicio A esta ejecutando en su entorno de ejecución el servicio B cae, el servicio A no debe interumpirse.

### Servicio sin estado

Un servicio no debe guardar ningún tipo de información, todos los parametros de entrada y lo que se espera de salida deben proveenir directamente de la documentación. 

### Componibilidad del servicio

Juntar servicios de bajo nivel para que sea un servicio de mayor nivel.

### Reutilización de servicios

El principal objetivo de SOA es construir servicios altamente reutilizables enfocado que se va a utilizar en varios aplicaciones y esto generar un valor ya que el mantenimiento sera más fácil de implementar. 



## Diferentes contratos de servicios

- OPEN API: Es el estándar para definir APIs RESTful
- Swagger: Implementa las plantillas que da OPEN API y lo adapta incluye interfaces, entre otras cosas, pueden aparecer más.

- WSDL: Es una especificación XML enfocada en SOAP
- WADL: Describre APIs RESTful en formato XML, es el menos usado.

## SOAP VS REST VS grcp

- SOAP: Es un estándar para el intercambio de información utilizando XML
- REST: En cambio rest es una propuesta para crear servicios web, aunque no tiene estándares de seguridad tan avanzados como SOAP.
- GRCP: Es un marco de comunicación diseñada por Google, los datos se serealizan en formatos binarios usando Protocol Buffer, convierte la infomación en formato binario lo que hace que trasmita más rápido y consuma menos ancho de banda que otros formatos como JSON o XML.
